```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Dataset = pd.read_csv("Preprocessed_ecommerce.csv")

Dataset
```

```python
Independent = Dataset[["Description", "Quantity", "UnitPrice"]]
Dependent = Dataset[["Best_Selling"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Independent, Dependent, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFClassifier = RandomForestClassifier()
RFClassifier.fit(x_train, y_train)
```

```python
RFImportance = RFClassifier.feature_importances_
RFImportance
```

```python
y_pred = RFClassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
RFcm = confusion_matrix(y_test, y_pred)
RFcm
```

```python
# We are creating report

from sklearn.metrics import classification_report
RFResult = classification_report(y_test, y_pred)
```

```python
print(RFResult)
```
